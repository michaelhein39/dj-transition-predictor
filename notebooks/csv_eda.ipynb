```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/meta/tracks_trunc.csv')
print(df.info())
print(df.describe())
```

```python
print(df['audio_sr'].describe())

plt.hist(df['audio_sr'], bins=20)
plt.title('Distribution of Audio Sample Rates')
plt.xlabel('Audio Sample Rate (Hz)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histograms
for col in ['bpm', 'audio_size', 'audio_sr', 'audio_length']:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col], bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
